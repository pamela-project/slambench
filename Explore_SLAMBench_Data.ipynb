```python
## For example to generate the pareto Accuracy / ExecutionTime

import pandas as pd
import matplotlib.pyplot as plt 

mycleandata = [ {'file': k, 'MaxATE':  x ['Summary']['algo']['AbsoluteError']['MAX'], 'FPS':  (1 / x ['Summary']['algo']['Duration_Frame']['MEAN'])} for (k,x) in list(data.items())[1:] ]

df = pd.DataFrame(mycleandata)
df.set_index("file")
dfx = df [ df["MaxATE"] < 0.5 ]
plt.scatter(1 / dfx["FPS"], dfx["MaxATE"])
```
